The Height is even worse, the first number are the feets and the second the inches

|Feets|meters|
|:----:|:-------:|
|1|0.3048|

|Inches|meters|
|:----:|:-------:|
|1|0.0254|

```python
def nba_height_fixer(h):
    
    if type(h) == str:
        h = h.split('-')

        return round(int(h[0]) * 0.3048 + int(h[1]) * 0.0254, 2)
        
    
    else:
        return np.nan
```

## Skewness

In probability theory and statistics, skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. The skewness value can be positive, zero, negative, or undefined.


<img src = https://upload.wikimedia.org/wikipedia/commons/c/cc/Relationship_between_mean_and_median_under_different_skewness.png>


**Fisher's coefficient of skewness**, represented by $\gamma_{1}$, is defined as:

$$\gamma_{1} = \frac{\mu_{3}}{\sigma^{3}}$$

where $\mu_3$ is the third moment around the mean and $\sigma$ is the standard deviation.

- If $\gamma_{1} > 0$ the distribution is positive or right skewed.

<img src = https://cdn.corporatefinanceinstitute.com/assets/Positively-Skewed-Distribution.png width = 500>

- If $\gamma_{1} <0$ the distribution is negative or left skewed.

<img src = https://cdn.corporatefinanceinstitute.com/assets/negatively-skewed-distribution-1024x530.png width = 500>

- If the distribution is symmetric, then we know that $\gamma_{1} = 0$. The converse is not true: it is a common mistake to ensure that if $\gamma_{1} = 0$ then the distribution is symmetric (which is false).


In `Python` we can use `scipy` to calculate [Fisher's coefficient of skewness](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.skew.html)

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set()
```

```python
sturges = lambda n: int(np.round(1 + np.log2(n)))
```

```python
stats.mode(data)[0][0]
```

```python
fig, ax = plt.subplots(1, 3, figsize = (18, 5))


# negative or left skewed.
np.random.seed(2021)
data = -1 * np.random.poisson(3, 5000) ** 2

ax[0].hist(data, bins = sturges(len(data)), color = 'skyblue')
ax[0].axvline(data.mean(), color = 'r', label = 'Mean', alpha = 0.9)
ax[0].axvline(np.median(data), color = 'g', label = 'Median', alpha = 0.9)
ax[0].axvline(stats.mode(data)[0][0], color = 'b', label = 'Mode', alpha = 0.9)
ax[0].legend(prop = {'size' : 10})
ax[0].set_title("Fisher's coefficient of skewness: {}".format(round(stats.skew(data), 2)))



#normal case
norm = np.random.normal(0, 1, 50000)
norm = pd.Series(norm)
norm.mode().median()


ax[1].hist(norm, bins = sturges(len(norm)), color = 'lightcoral')
ax[1].axvline(norm.mean(), color = 'r', label = 'Mean', alpha = 0.9)
ax[1].axvline(norm.median(), color = 'g', label = 'Median', alpha = 0.9)
ax[1].axvline(norm.mode().median(), color = 'b', label = 'Mode', alpha = 0.9)
ax[1].legend(prop = {'size' : 10})
ax[1].set_title("Fisher's coefficient of skewness: {}".format(round(stats.skew(norm), 2)))



# positive or right skewed.
data *= -1
ax[2].hist(data, bins = sturges(len(data)), color = 'limegreen')
ax[2].axvline(data.mean(), color = 'r', label = 'Mean', alpha = 0.9)
ax[2].axvline(np.median(data), color = 'g', label = 'Median', alpha = 0.9)
ax[2].axvline(stats.mode(data)[0][0], color = 'b', label = 'Mode', alpha = 0.9)
ax[2].legend(prop = {'size' : 10})
ax[2].set_title("Fisher's coefficient of skewness: {}".format(round(stats.skew(data), 2)));
```

Key insight:

- A value on its own is dificult to interpret
- A value relative to its distriibution is easier to interpret


### Z-transform

$$z_i = \frac{x_i - \overline{x}}{\sigma_x}$$


- Mean-center: Substrac the average from each individual value.
- Variance-normalize: Divide by the standard deviation.


The units are standard deviation away from the mean of the distribution.

```python
fig, ax = plt.subplots(1, 2, figsize = (16, 5))

#generate the samples
heightsN = np.random.randn(1000)
heights = 0.22 * heightsN + 1.68


#plot histogram color = [.7, .3, 0.6]
ax[0].hist(heights, bins = sturges(len(heightsN)), color = 'skyblue')
ax[1].hist(heightsN, bins = sturges(len(heights)), color = 'limegreen')


#customization

ax[0].set_xlabel('Height in m', fontsize = 15)
ax[1].set_xlabel('Height in $\\sigma$', fontsize = 15);
```

**Assumption**

- Mean and standard deviation are valid descriptions of the distribution's central tendency and dispersion.

```python
## create data

data = np.random.poisson(3, 1000)

plt.hist(data, bins = sturges(len(data)));
```

$$\sigma^2 = \frac{\sum (x_i - \overline{x})^2}{N - 1}$$

```python
# compute the mean and std
datamean = data.mean()
datastd  = data.std(ddof = 1) #Delta degrees of Freedom

fig, ax = plt.subplots(figsize = (13, 8))
ax.plot(data, linestyle = '', marker = 'o', markersize = 3)


#customization
ax.set_xlabel('Data index', fontsize = 15)
ax.set_ylabel('Data value', fontsize = 15)
ax.set_title('Mean = {}; std = {}'.format(np.round(datamean, 2), np.round(datastd, 2)), fontsize = 15);
```

```python
## now for z-scoring

# z-score is data minus mean divided by stdev
dataz = (data - datamean) / datastd

# can also use Python function
dataz = stats.zscore(data)

# compute the mean and std
dataZmean = np.mean(dataz)
dataZstd  = np.std(dataz, ddof = 1)

fig, ax = plt.subplots(figsize = (13, 8))
ax.plot(dataz, linestyle = '', marker = 'o', markersize = 3, color = 'limegreen')


#customization
ax.set_xlabel('Data index', fontsize = 15)
ax.set_ylabel('Data value', fontsize = 15)
ax.set_title('Mean = {}; std = {}'.format(np.round(dataZmean, 2), np.round(dataZstd, 2)), fontsize = 15);
```

```python
fig, ax = plt.subplots(figsize = (13, 8))

ax.plot(data, dataz,'-ks')
ax.set_xlabel('Original', fontsize = 15)
ax.set_ylabel('Z-transformed', fontsize = 15);
```

## Max-Min Scaling

- How to scale your data to any arbitrary range (typically $[0, 1]$)

$$\tilde{x}_i = \frac{x_i - \min x}{\max x - \min x}$$

```python
## create some data
N = 42
data = np.log(np.random.rand(N)) * 234 + 934

# get min and max
dataMin = min(data)
dataMax = max(data)

# now min-max scale
dataS = (data - dataMin) / (dataMax - dataMin)


# now plot
fig,ax = plt.subplots(1, 2, figsize = (15, 8))

ax[0].plot(1 + np.random.randn(N) / 20, data, 'ks')
ax[0].axhline(dataMin, color = 'g', alpha = 0.7, label = 'min = {}'.format(np.round(dataMin, 2)))
ax[0].axhline(dataMax, color = 'r', alpha = 0.7, label = 'max = {}'.format(np.round(dataMax, 2)))
ax[0].set_xlim([0, 2])
ax[0].set_xticks([])
ax[0].set_ylabel('Original data scale')
ax[0].set_title('Original data')
ax[0].legend()

ax[1].plot(1 + np.random.randn(N) /20 ,dataS,'ks')
ax[1].axhline(max(dataS), color = 'r', alpha = 0.7, label = 'max = {}'.format(max(dataS)))
ax[1].axhline(min(dataS), color = 'g', alpha = 0.7, label = 'min = {}'.format(min(dataS)))
ax[1].set_xlim([0, 2])
ax[1].set_xticks([])
ax[1].set_ylabel('Unity-normed data scale')
ax[1].set_title('Scaled data')
ax[1].legend();
```

```python
fig,ax = plt.subplots(1,2,figsize=(13,6))

ax[0].violinplot(data)
ax[0].axhline(dataMax, color = 'r', alpha = 0.7, label = 'max = {}'.format(np.round(dataMax, 2)))
ax[0].axhline(dataMin, color = 'g', alpha = 0.7, label = 'min = {}'.format(np.round(dataMin, 2)))
ax[0].set_xlim([0,2])
ax[0].set_xticks([])
ax[0].set_ylabel('Original data scale')
ax[0].set_title('Original data')
ax[0].legend()


ax[1].violinplot(dataS)
ax[1].axhline(max(dataS), color = 'r', alpha = 0.7, label = 'max = {}'.format(max(dataS)))
ax[1].axhline(min(dataS), color = 'g', alpha = 0.7, label = 'min = {}'.format(min(dataS)))
ax[1].set_xlim([0,2])
ax[1].set_xticks([])
ax[1].set_ylabel('Unity-normed data scale')
ax[1].set_title('Scaled data')
ax[1].legend();
```

```python
## show that scaling doesn't affect the relative values

fig, ax = plt.subplots(figsize = (13, 8))

ax.plot(data, dataS,'-ks')
ax.set_xlabel('Original', fontsize = 15)
ax.set_ylabel('Scaled', fontsize = 15);
```

scale to a $[a, b]$:

$$x^* = a + \tilde{x} (b - a)$$

```python
## any abitrary data range

# step 1 is to [0,1] normalize as above

# step 2:
newMin = 4
newMax = 8.7

dataSS = dataS * (newMax - newMin) + newMin

# test it!
print([min(dataSS), max(dataSS)])

#and plot it
fig,ax = plt.subplots(1, 2, figsize = (13, 6))

ax[0].violinplot(dataS)
ax[0].axhline(max(dataS), color = 'r', alpha = 0.7, label = 'max = {}'.format(max(dataS)))
ax[0].axhline(min(dataS), color = 'g', alpha = 0.7, label = 'min = {}'.format(min(dataS)))
ax[0].set_xlim([0, 2])
ax[0].set_xticks([])
ax[0].set_ylabel('Unity-normed data scale')
ax[0].set_title('Scaled data')
ax[0].legend()

ax[1].violinplot(dataSS)
ax[1].axhline(newMax, color = 'r', alpha = 0.7, label = 'max = {}'.format(newMax))
ax[1].axhline(newMin, color = 'g', alpha = 0.7, label = 'min = {}'.format(newMin))
ax[1].set_xlim([0, 2])
ax[1].set_xticks([])
ax[1].set_ylabel('New scale')
ax[1].set_title('New Scaled Data')
ax[1].legend();
```

# What are Outliers?

**How to deal with ouliers**

- Identify outliers and romove them from the data prior to any analyses.
    - Assumption: Outliers are noises or otherwise invalid.
    
    
- Leave outliers in and use robust methods that attenuate the negative impact of the outliers in the results.
    - Assumption: Outliers are unusual but valid data.

## Reomoving outliers Z-score method

$$z_i = \frac{x_i - \overline{x}}{\sigma_x}$$

Algorithm:

1. Convert data to z-score.
2. A data point is a outlierif ti exceed some standard deviation threshold (often 3).

```python
##seed
np.random.seed(42)

## create some data

N = 40
data = np.random.randn(N)


data[data < -1] = data[data < -1] + 2
data[data > 2] = data[data > 2] ** 2; # try to force a few outliers
data[21] = abs(data[21]) + 4
data[27] = abs(data[27]) + 3.2
data = data * 200 + 50  # change the scale for comparison with z


# convert to z
dataZ = (data - np.mean(data)) / np.std(data)


#### specify the z-score threshold
zscorethresh = 3



# plot the data
fig,ax = plt.subplots(2, 1, figsize=(13, 8))

ax[0].plot(data, 'k^', markerfacecolor = 'w', markersize = 12)
ax[0].set_xticks([])
ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Orig. scale')

# then plot the zscores
ax[1].plot(dataZ, 'k^', markerfacecolor = 'w', markersize = 12)
ax[1].plot([0,N], [zscorethresh,zscorethresh],'r--')
ax[1].set_xlabel('Data index')
ax[1].set_ylabel('Z distance');
```

```python
## identify outliers

# find 'em!
outliers = np.where(abs(dataZ) > zscorethresh)[0]#indices where the condition is met

# and cross those out
ax[0].plot(outliers, data[outliers], 'x', color = 'r', markersize=20)
ax[1].plot(outliers, dataZ[outliers], 'x', color = 'r', markersize=20)

fig
```

Algorithm:

1. Convert data to z-score.
2. A data point is a outlierif ti exceed some standard deviation threshold (often 3).
3. Remove outliers and repeat until no more outliers.

```python
np.array([False, False, False]).sum()
```

```python
## iterative method

# pick a lenient threshold just for illustration
zscorethresh = 2.0
dataZ = (data - np.mean(data)) / np.std(data)


fig, ax = plt.subplots(figsize = (13, 8))

colorz = 'brkmg'
numiters = 0 # iteration counter
while True:
    
    # convert to z
    datamean = np.nanmean(dataZ)
    datastd  = np.nanstd(dataZ)
    dataZ = (dataZ - datamean) / datastd
    
    # find data values to remove
    toremove = np.abs(dataZ) > zscorethresh
    
    # break out of while loop if no points to remove
    if sum(toremove) == 0:
        break
    
    else:
        # otherwise, mark the outliers in the plot
        ax.plot(np.where(toremove)[0], dataZ[toremove], colorz[numiters] + 'x', markersize = 12)
        ax.plot([0, len(dataZ)], [zscorethresh, zscorethresh], '--g')
        ax.plot([0, len(dataZ)], [-zscorethresh, -zscorethresh], '--g')
        dataZ[toremove] = np.nan
    
    # replot
    ax.plot(dataZ, 'k^', markersize = 12, markerfacecolor = colorz[numiters], label = 'iteration {}'.format(1 + numiters))
    numiters += 1

ax.set_xticks([])
ax.set_ylabel('Z-score')
ax.set_xlabel('Data index')
ax.legend()
plt.show()

#### the data points to be removed
removeFromOriginal = np.where(np.isnan(dataZ))[0]
print(removeFromOriginal)
```

## Modified Z-score Method 

$$M_i = \frac{Q_{75}(x_i - \tilde{x})}{MAD}$$


$$MAD = median(|x_i - \tilde{x}|)$$

$$\tilde{x} = median(x)$$

```python
from statsmodels import robust
```

```python
data = np.random.poisson(3, 5000) ** 2

fig, ax = plt.subplots(figsize = (13, 8))

ax.hist(data, bins = sturges(len(data)), color = 'skyblue');
```

```python
fig, ax = plt.subplots(2, 1, figsize = (13, 8))

ax[0].hist(data, bins = sturges(len(data)), color = 'skyblue')
ax[0].axvline(data.mean(), color = 'red', alpha = 0.7)
ax[0].axvline(data.mean() + data.std(), color = 'purple', alpha = 0.7)
ax[0].axvline(data.mean() - data.std(), color = 'purple', alpha = 0.7)

ax[1].hist(data, bins = sturges(len(data)), color = 'skyblue')
ax[1].axvline(np.median(data), color = 'green', alpha = 0.7)
ax[1].axvline(np.median(data) + robust.mad(data), color = 'red', alpha = 0.7)
ax[1].axvline(np.median(data) - robust.mad(data), color = 'red', alpha = 0.7);
```

```python
## modified Z for non-normal distributions

# compute modified z
dataMed = np.median(data)
dataMAD = robust.mad(data)

dataMz = stats.norm.ppf(.75) * (data - dataMed) / dataMAD
zscorethresh = 3.0

# plot the data
fig, ax = plt.subplots(2, 1, figsize = (13, 8))

ax[0].plot(data, 'k^', markerfacecolor = 'w', markersize = 12)#, markevery = 20)
ax[0].set_xticks([])
ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Orig. scale')

# then plot the zscores
ax[1].plot(dataMz, 'k^', markerfacecolor = 'w', markersize = 12)#, markevery = 20)
ax[1].plot([0, len(data)], [zscorethresh, zscorethresh], '--r')
ax[1].set_xlabel('Data index')
ax[1].set_ylabel('Median dev. units (Mz)');
```

## Multivariable Detection

```python
## create some data

N = 40

# two-dimensional data
d1 = np.exp(-abs(np.random.randn(N) * 3))
d2 = np.exp(-abs(np.random.randn(N) * 5))
datamean = [np.mean(d1), np.mean(d2)]



# plot the data
fig, ax = plt.subplots(figsize = (13, 8))

ax.plot(d1, d2, 'ko')
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Variable x')
ax.set_ylabel('Variable y')

# plot the multivariate mean
ax.plot(datamean[0], datamean[1], 'kp', markerfacecolor = 'g', markersize = 15);
```

$$d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$$

```python
## create some data
N = 40

# two-dimensional data
d1 = np.exp(-abs(np.random.randn(N) * 3))
d2 = np.exp(-abs(np.random.randn(N) * 5))
datamean = [np.mean(d1), np.mean(d2)]


# compute distance of each point to the mean
ds = np.zeros(N)
for i in range(N):
    ds[i] = np.sqrt((d1[i] - datamean[0]) ** 2 + (d2[i] - datamean[1]) ** 2)
    

# convert to z (don't need the original data)
ds = (ds - np.mean(ds)) / np.std(ds)





# plot the data
fig, ax = plt.subplots(1, 2, figsize = (13, 8))

ax[0].plot(d1, d2, 'ko', markerfacecolor = 'k')
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].set_xlabel('Variable x')
ax[0].set_ylabel('Variable y')

# plot the multivariate mean
ax[0].plot(datamean[0], datamean[1], 'kp', markerfacecolor = 'g', markersize = 15)

# then plot those distances
ax[1].plot(ds, 'ko', markerfacecolor = [.7, .5, .3], markersize = 12)
ax[1].set_xlabel('Data index')
ax[1].set_ylabel('Z distance')
```

## Removing Outliers by Data Trimming

1. Sort the mean-centered data
2. Remove the most extreme $k$ values or the most extreme $k\%$

```python
np.random.seed(1)
## create some data

N = 40
data = np.random.randn(N)
data[data < -2] = -data[data <- 2] ** 2
data[data > 2]  =  data[data > 2] ** 2

# also need the mean-centered data
dataMC = data - np.mean(data)

# and plot them
fig,ax = plt.subplots(figsize = (13, 6))
ax.plot(data, 'k^', markerfacecolor = 'y', markersize = 12)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Data index')
ax.set_ylabel('Data value');
```

```python
dataMCcopy = dataMC.copy()
dataMCcopy.sort()

fig, ax = plt.subplots(figsize = (13, 6))
ax.plot(abs(dataMCcopy), 'k^', markerfacecolor = 'y', markersize = 12)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Data index')
ax.set_ylabel('Data value');
```

```python
## option 1: remove k% of the data

# percent of "extreme" data values to remove
trimPct = 5 # in percent

# identify the cut-off (note the abs())
datacutoff = np.percentile(abs(dataMC), 100 - trimPct)

# find the exceedance data values
data2cut = np.where(abs(dataMC) > datacutoff)[0]

# and mark those off
fig, ax = plt.subplots(figsize = (13, 6))

ax.plot(data, 'k^', markerfacecolor = 'y', markersize = 12)
ax.plot(data2cut, data[data2cut], 'rx', markersize = 15)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Data index')
ax.set_ylabel('Data value');
```

```python
## option 2: remove k most extreme values

# number of "extreme" data values to remove
k2remove = 3  # in number

# find the exceedance data values
datasortIdx = np.argsort(abs(dataMC))[::-1]
data2cut = datasortIdx[:k2remove]

# and mark those off
ax.plot(data2cut, data[data2cut], 'go', markersize = 15, alpha = .5)
fig
```
