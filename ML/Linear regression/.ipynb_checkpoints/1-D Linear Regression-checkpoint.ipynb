# Linear Regression (One dimensional case)


The idea under the linear regression is looking for the line model that fit better our data points.

$$\hat{y} = \alpha + \beta x$$

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact
%matplotlib inline
```

```python
sns.set()
```

```python
np.random.seed(1995)
```

```python
x = stats.norm(1.5, 2.5).rvs(100)
res = stats.norm(0, 0.3).rvs(100)
y_pred = 5 + 0.3 * x
y_real = 5 + 0.3 * x + res
```

```python
fig, axs = plt.subplots(1, 2, figsize = (13, 6))

axs[0].plot(x, y_real, 'or')

axs[1].plot(x, y_real, 'or')
axs[1].plot(x, y_pred);
```

We'll use the data to looking for $\alpha$ and $\beta$ statistically significant for our model. With this equation we can realise predictions, but how we can found the parameters?. for this we need focus to the error.

$$e_i = (y_i - \hat{y}_i (x_i))$$

```python
fig, axs = plt.subplots(1, 2, figsize = (13, 6))


axs[0].plot(x, y_real, 'or')
axs[0].plot(x, y_pred)

axs[1].plot(x, y_real, 'or')
axs[1].plot(x, y_pred)

for X, Y, y in zip(x, y_real, y_pred):
    axs[1].plot([X, X], [Y, y], linestyle = ':', color = 'y')
```


So, we are going to minimize these errors, that is, we have to minimize the following function.


$$S(\alpha, \beta) =  \sum e_{i}^{2} = \sum (y_i - \hat{y}(x_i))^2 = \sum (y_i - (\alpha + \beta x_i))^2$$


We will call this function the cost function.

- why do we need to square the error, why is the absolute value not used?



```python
fig, axs = plt.subplots(1, 2, figsize = (13, 6))

x2 = np.linspace(-2, 2, 1000)

axs[0].plot(x2, x2 ** 2)

axs[1].plot(x2, np.abs(x2));
```

Naturally to minimize we are going to calculate the gradient of this function.


- $\frac{\partial S}{\partial \alpha} = -2 \sum (y_i - \alpha - \beta x_i) = 0$
- $\frac{\partial S}{\partial \beta} = -2 \sum (y_i - \alpha - \beta x_i) x_i = 0$


$$
\left\{
\begin{matrix}
(\sum x_i ) \alpha + (\sum x_{i}^{2}) \beta = \sum y_i x_i\\
n \alpha + (\sum x_i ) \beta = \sum y_i\\
\end{matrix}
\right.
$$

## Exercise

Find the $\alpha$ and $\beta$ parameters and implement the model

- $\alpha + \bar{x} \beta = \bar{y}$
	- $\alpha = \bar{y} - \bar{x} \beta$


- $(\sum x_i ) (\bar{y} - \bar{x} \beta) + (\sum x_{i}^{2}) \beta = \sum y_i x_i$



$$\bar{y} \sum x_i - \beta \bar{x} \sum x_i + \beta \sum x_{i}^{2} = \sum y_i x_i$$

$$\beta (\sum x_{i}^{2} - \bar{x} \sum x_i) = \sum y_i x_i - \bar{y} \sum x_i$$

$$\beta = \frac{\sum y_i x_i - \bar{y} \sum x_i}{\sum x_{i}^{2} - \bar{x} \sum x_i}$$

$$\alpha = \bar{y} - \bar{x} \beta$$

```python
class LinearRegression:
    
    def __init__(self, x, y):
        
        n = len(x)
        
        self.beta  = ((x * y).sum() - y.mean() * x.sum()) / ((x ** 2).sum() - x.mean() * x.sum())
        self.alpha = y.mean() - x.mean() * self.beta
        
        
    def predict(self, x):
        
        return self.beta * x + self.alpha
```

```python
model = LinearRegression(x, y_real)
```

```python
y_pred = model.predict(x)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (13, 6))

axs[0].plot(x, y_real, 'or')

axs[1].plot(x, y_real, 'or')
axs[1].plot(x, y_pred);
```

$$\beta = \frac{ n \left ( \sum y_i x_i - \sum y_i \sum x_i \right)}{n \left(\sum x_{i}^{2} - (\sum x_i)^2 \right)} $$

$$\beta = \frac{cov(x, y)}{var(x)}$$

```python
np.cov(x, y_real)
```

```python
class LinearRegression:
    
    def __init__(self, x, y):
        
        n = len(x)
        
        self.beta  = np.cov(x, y)[0, 1] / x.var()
        self.alpha = y.mean() - x.mean() * self.beta
        
        
    def predict(self, x):
        
        return self.beta * x + self.alpha
```

```python
model = LinearRegression(x, y_real)
```

```python
y_pred = model.predict(x)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (13, 6))

axs[0].plot(x, y_real, 'or')

axs[1].plot(x, y_real, 'or')
axs[1].plot(x, y_pred);
```

## How good is our model?

```python
data = pd.DataFrame({"x" : x, "Real Data" : y_real, "Prediccion" : y_pred})
```

```python
data.head()
```

```python
def plot_reg(model, mean, point, SSR, SST, SSM):
    
    fig, ax = plt.subplots(figsize = (13, 8))
    
    
    ax.plot(data['x'], data['Real Data'], 'or', label = 'Real Observation')
    
    
    if model:
        ax.plot(data['x'], data['Prediccion'], label = 'Model')
        
    
    if mean:
        ax.axhline(data['Real Data'].mean(), linestyle = ':', color = 'g', alpha = 0.7, label = 'Mean', lw = 4)
    
    if point:
        ax.plot([-1], [6.5], '*y', markersize = 20)
        
    if SSR:
        ax.plot([-1, -1], [5 + 0.3 * -1, 6.5], '--', color = 'blue', label = 'SSR')
        
    if SST:
        ax.plot([-1, -1], [data['Real Data'].mean(), 6.5], '--', color = 'purple', label = 'SST')
    
    if SSM:
        ax.plot([-1, -1], [data['Real Data'].mean(), 5 + 0.3 * -1], '--', color = 'orange', label = 'SSM')
        
    
    ax.set_xlabel('X', fontsize = 15)
    ax.set_xlabel('Y', fontsize = 15)
    ax.set_title('Model vs Reality', fontsize = 15)
    ax.legend(prop = {'size' : 13})
    
    
```

```python
interact(plot_reg, model = False, mean = False, point = False, SSR = False, SST = False, SSM = False);
```

- $SSR = \sum (y_i - \hat{y}_i)^2$
- $SST = \sum (y_i - \bar{y})^2$
- $SSM = \sum (\hat{y}_i - \bar{y})^2$


Note that

$$y_i - \bar{y} = (y_i - \hat{y}_i) + (\hat{y}_i - \bar{y})$$

Now squere that.

$$(y_i - \bar{y})^2 = (y_i - \hat{y}_i)^2 + (\hat{y}_i - \bar{y})^2 + 2 (y_i - \hat{y}_i)(\hat{y}_i - \bar{y})$$

$$SST = SSR + SSM + 2 \sum (y_i - \hat{y}_i)(\hat{y}_i - \bar{y})$$


Then we see that $2 \sum (y_i - \hat{y}_i)(\hat{y}_i - \bar{y}) = 0$ for this I think about the following:

$$(y_i - \hat{y}_i)(\hat{y}_i - \bar{y}) = (y_i - \hat{y}_i)\hat{y}_i -  (y_i - \hat{y}_i)\bar{y}$$
$$\sum (y_i - \hat{y}_i)(\hat{y}_i - \bar{y}) = \sum (y_i - \hat{y}_i)\hat{y}_i -  \sum (y_i - \hat{y}_i)\bar{y}$$

And remember that the model found $\beta$ and $\alpha$ assuming:

$$\frac{\partial S}{\partial \alpha} = -2 \sum (y_i - \alpha - \beta x_i) = -2 \sum (y_i - \hat{y}_i)=0$$
$$\frac{\partial S}{\partial \beta} = -2 \sum (y_i - \alpha - \beta x_i) x_i = -2 \sum (y_i - \hat{y}_i) x_i=0$$


Hence

$$\sum (y_i - \hat{y}_i)(\hat{y}_i - \bar{y}) = 0$$


And we conclude that:

$$SST = SSR + SSM$$


<hr>
$$1 = \frac{SSR}{SST} + \frac{SSM}{SST}$$
<hr>
$$1 - \frac{SSR}{SST} = \frac{SSM}{SST}$$
<hr>
$$R^2 = \frac{SSM}{SST}$$
<hr>
$$R^2 = 1 - \frac{SSR}{SST} = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}$$



### Exercise

Implement $R^2$

```python
class LinearRegression:
    
    def __init__(self, x, y):
        
        n = len(x)
        
        self.beta  = np.cov(x, y)[0, 1] / x.var()
        self.alpha = y.mean() - x.mean() * self.beta
        
        SSR = ((y - self.predict(x)) ** 2).sum()
        SST = ((y - y.mean()) ** 2).sum()
        
        self.R2 = 1 - (SSR / SST)
        
    def predict(self, x):
        
        return self.beta * x + self.alpha
```

```python
model = LinearRegression(x, y_real)
```

```python
y_pred = model.predict(x)
```

```python
fig, axs = plt.subplots(1, 2, figsize = (13, 6))

axs[0].plot(x, y_real, 'or')

axs[1].plot(x, y_real, 'or')
axs[1].plot(x, y_pred)
axs[1].set_title('$R^2$ = {}'.format(model.R2));
```

That seems to change, but it doesn't work at all better, what if the model was something like that?

$$y = \alpha e^{k x}$$



If that were the case we can linearize by doing the following

$$\ln(y) = \ln(\alpha e^{k x})$$

$$\ln(y) = \ln(\alpha) + \ln(e^{k x})$$
$$\ln(y) = \ln(\alpha) + k x$$

```python
fig, ax = plt.subplots(figsize = (13, 8))

ax.plot(np.log(X), np.log(Y), "ro")
ax.set_xlabel("horsepower", fontsize = 15)
ax.set_ylabel("Milles per gal", fontsize = 15)
ax.set_title("HP vs MPG", fontsize = 15);
```

```python
model = LinearRegression(np.log(X), np.log(Y))
```

```python
fig, ax = plt.subplots(figsize = (13, 8))

ax.plot(np.log(X), np.log(Y), "ro", label = 'Real')
ax.plot(np.log(X), model.predict(np.log(X)), 'bo', label = 'model')
ax.set_xlabel("horsepower", fontsize = 15)
ax.set_ylabel("Milles per gal", fontsize = 15)
ax.set_title("HP vs MPG", fontsize = 15)
ax.legend(prop = {'size' : 15});
```

```python
model.alpha, model.beta, model.R2
```

```python
fig, ax = plt.subplots(figsize = (13, 8))

ax.plot(X, Y, "ro", label = 'Real')
ax.plot(X, np.exp(model.predict(np.log(X))), 'bo', label = 'model')
ax.set_xlabel("horsepower", fontsize = 15)
ax.set_ylabel("Milles per gal", fontsize = 15)
ax.set_title("HP vs MPG", fontsize = 15)
ax.legend(prop = {'size' : 15});
```

```python
def func(x):
    return np.exp(model.alpha + model.beta * np.log(x))
```

```python
fig, ax = plt.subplots(figsize = (13, 8))

ax.plot(X, Y, "ro", label = 'Real')
ax.plot(X, func(X), 'bo', label = 'model')
ax.set_xlabel("horsepower", fontsize = 15)
ax.set_ylabel("Milles per gal", fontsize = 15)
ax.set_title("HP vs MPG", fontsize = 15)
ax.legend(prop = {'size' : 15});
```

### Exercise, Moore's law


Moore's law is the observation that the number of transistors in a dense integrated circuit (IC) doubles about every two years.

```python
data = pd.read_csv('moore.csv', delimiter = '\t', names = ['Model', 'Transistors', 'Year', 'Company', 'TS', 'area'])
```

```python
data.head()
```

```python
data.info()
```

```python
data['Year'].values
```

```python
data['Transistors'].values
```

```python
data['Year'] = data['Year'].apply(lambda x: x.split('[')[0])
data['Transistors'] = data['Transistors'].apply(lambda x: x.split('[')[0])
```

```python
data['Year'].values
```

```python
data['Transistors'].values
```

```python
data['Transistors'] = data['Transistors'].apply(lambda x: x.replace(',', ''))
```

```python
data['Transistors'].values
```

```python
def clean_func(x):
    aux = ''
    
    for letter in x:
        if letter.isnumeric():
            aux += letter
                
    return aux
```

```python
data['Transistors'] = data['Transistors'].apply(clean_func)
```

```python
data['Transistors'].values
```

```python
X, Y = data['Year'].astype(int).values, data['Transistors'].apply(int).values
```

```python
fig, ax = plt.subplots(figsize = (13, 8))

ax.plot(X, Y, "ro")
ax.set_xlabel("Year", fontsize = 15)
ax.set_ylabel("Transistors", fontsize = 15)
ax.set_title("Moore's law", fontsize = 15);
```

This seems to be an exponential curve, which is just what Moore's law says and therefore our hypothesis


$$y = \beta e^{kx}$$


If we apply natural logarithm to both sides

$$\ln(y) = \ln(\beta e^{kx})$$
$$\ln(y) = \ln(\beta) +  \ln(e^{kx})$$
$$\ln(y) = \ln(\beta) +  kx$$


**Use the linear regression model to find this line**


Now how often does the number of transistors double?

$$y = \beta e^{kx_1}$$

$$2y = \beta e^{kx_2} = 2 \beta e^{kx_1}$$

Now

$$2 \beta e^{kx_1} = e^{(\ln(2))} \beta e^{kx_1}$$

$$\beta e^{kx_1 + \ln(2)}$$

So

$$\beta e^{kx_2} = \beta e^{kx_1 + \ln(2)}$$

$$kx_2 = kx_1 + \ln(2)$$

$$x_2 = x_1 + \frac{\ln(2)}{k}$$


This means that the transistors must be doubled at a time $\frac{\ln(2)}{k}$

```python
Y = np.log(Y)
```
